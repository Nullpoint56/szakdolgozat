# PDF processing pipeline
## Process steps
1. Create a searchable text layer on PDF documents using OCRMyPDF library and save the searchable PDF files.
2. Loading the PDf files, extracting text from them using PyMuPDF and storing them in a large array.
3. Chunking data and transforming the chunks into embeddings using Jina embedder model.

# Problems
Issues emerged in the process of chunking the data. Specifically, we didn't capture the chapters and other structural data from the PDF which are crucial for splitting the document into chunks that will form strong clusters in the vector space and ensure that related data will remain related in the vector database.


```python
import subprocess

# Example usage
input_pdf = "/mnt/d/python projects/RAG_chatbot_szakdolgozat/data/source_documents/PublicWaterMassMailing.pdf"
output_pdf = "/mnt/d/python projects/RAG_chatbot_szakdolgozat/data/processed_documents/PublicWaterMassMailing.pdf"

def process_pdf(input_pdf_path, output_pdf_path):
    try:
        # Apply OCR with ocrmypdf, allowing it to automatically determine
        # which pages need OCR. This is ideal for mixed-content PDFs.
        subprocess.run(["ocrmypdf", input_pdf_path, output_pdf_path, "--redo-ocr", ], check=True, )
        print(f"Processed PDF saved as: {output_pdf_path}")
    except subprocess.CalledProcessError as e:
        print("Error during the OCR process:", e)
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


process_pdf(input_pdf, output_pdf)
```
