# Loading PDF documents for RAG
PDF documents are untagged documents, meaning they contain instructions on the placement of content elements, rather than structuring tags and content. This makes them unreadable for machines. For parsing these documents two approaches emerged: 
- Rule based: Serializes characters in a PDF into a long sequence without document structure information and then segment this sequence using segmentation rules.
- Deep learning based: Uses machine learning models for text position and recognition, physical document object detection, Cross-column and cross-page trimming, reading order determination, table structure recognition and document logical structure recognition. After these elements have been parsed, the contents are merged.
The rule based approach has several issues:
1. It cannot recognize the boundary of paragraphs and tables.
2. It cannot recognize the structure within a table.
3. It cannot recognize the reading order of the content.
This can significantly reduce the efficiency of document processing and negatively hinder the results of our RAG system. However, deep learning based parsing is developed to eliminate these issues.
